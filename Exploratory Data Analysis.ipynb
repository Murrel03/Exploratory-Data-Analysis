```python
#Checking whether there are errors or not

correlation = df.corr() 
```

```python
sns.set(style="darkgrid")
```

### Heatmap

```python
'''heatmap is a two dimensional visual representation of data where values are encoded in colors,
delivering a convenient, insightful view of information'''

sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, 
            annot=True)
```

### Pairplot

```python
'''Pairplot visualizes the given data to find the relationship between them 
where the variables can be continous or categorical'''

sns.pairplot(df)
```
